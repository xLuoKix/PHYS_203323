```python
import numpy as np
# Copy your Particle class here
class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
    def KE(self):
        return 1/2*(self.mass*((self.vx)**2 + (self.vy)**2 + (self.vz)**2))
    def distance(self, other):
        return np.sqrt((self.x-other.x)**2 + (self.y-other.y)**2 + (self.z-other.z)**2)
    def move(self, dt):
        self.x = self.x + self.vx*dt
        self.y = self.y + self.vy*dt
        self.z = self.z + self.vz*dt

#class Charged_Particle(Particle):
#    def charge(self, Pos, Neg):
#        return(
# part 5
class ChargedParticle(Particle):
    
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        # Add a new attribute for electric charge
        self.charge = charge

    def interact(self, other):
        if self.distance(other) <= 0.25:
            if (self.charge > 0 and other.charge > 0) or (self.charge < 0 and other.charge < 0):
                self.vx *= (-1 * (self.charge + other.charge) * (self.mass/other.mass))
                self.vy *= (-1 * (self.charge + other.charge) * (self.mass/other.mass))
                self.vz *= (-1 * (self.charge + other.charge) * (self.mass/other.mass))
                print("repel")
            elif (self.charge > 0 and other.charge < 0) or (self.charge < 0 and other.charge > 0):
                self.vx = 0
                self.vy = 0
                self.vz = 0
                self.charge = 0
                other.vx = 0
                other.vy = 0
                other.vz = 0
                other.charge = 0
                print("merge")
        #else:
            #print(self.distance(other))
p1 = ChargedParticle(10,-5,-5,-5,1,1,1, 0.5)
p2 = ChargedParticle(10,5,5,5,-1,-1,-1, -0.5)
print(p1.KE(), p2.KE())
print(p1.distance(p2))

print(p1.x, p1.y, p1.z)
p1.move(13)
print(p1.x, p1.y, p1.z)
p2.move(5)
```
